## Pie chart

```python
#pie char >> To understand the percent distribution of data

explode = (0.0, 0.0, 0.1)
attribute = ["C", "Python", "Java"]
quantity = [20, 100, 40]

plt.pie(quantity, labels=attribute, autopct='%1.1f%%', explode=explode, shadow=True)
plt.show()
```
