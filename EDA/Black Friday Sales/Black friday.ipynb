```python
df = pd.concat([df_train, df_test], ignore_index=True)
df.head()
```

# Fixing categorical values (by converting cate.. to numerical)

```python
pd.get_dummies(df['Gender'])
df.info()
```

```python
pd.get_dummies(df['Gender'], drop_first=1)
```

#### Above method will be lengthy. Create new column, then add to original df. 

#### Another way. (Change directly in the df)
There are 100 ways to change

## Use map method map({ key:value, key:value})

```python
df.info()
```

```python
df['Gender'] = df["Gender"].map({'F':0,'M':1})
```

```python
df.head()
```

```python
# Also
df['Gender'] = pd.get_dummies(df['Gender'])
```

```python
# Handle categorical feature: age
df['Age'].unique()
```

```python
df['Age'] = df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
df.head()
```

```python
# Fixing categorical feature: City category
df['City_Category'].unique()
```

```python
df_city = pd.get_dummies(df['City_Category'],drop_first=True)
df_city.head()
```

```python
df = pd.concat([df,df_city],axis=1)
df.head()
```

### Drop_first: If there are k categories, k-1 categories are sufficient to represent all categories

```python
# Now, there is no need of City_category column. delete it df.drop()
df.drop('City_Category',axis=1,inplace=True)
df.head()
```

### Replacing missing values

```python
df['Product_Category_2'].unique()
```

```python
df['Product_Category_2'].mode().reset_index()
```

```python
replace_with = df['Product_Category_2'].mode()[0]
replace_with
```

```python
df['Product_Category_2'] = df['Product_Category_2'].fillna(replace_with)
```

```python
df.head()
```

```python
df['Product_Category_3'] = df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```

```python
df.head()
```

```python
df['Stay_In_Current_City_Years'].info()
```

```python
# Convert object form in number to integer

df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].replace('4+','4')
```

```python
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].astype(int)
```

```python
df.info()
```

```python
df['B'] = df['B'].astype(int)
df['C'] = df['C'].astype(int)
```
