```python
df.isnull().sum()
```

```python
# Find missing value column
[i for i in df.columns if df[i].isnull().sum() > 0]
```

```python
# Highlights missing values on the map corresponding to lable

sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country
```

```python
# merge tables (left join)
final_df = pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.head()
```

```python
final_df.Country.value_counts()
```

```python
final_df.Country.value_counts().values
```

```python
# Take country names and values
country_values = final_df.Country.value_counts().values
country_names = final_df.Country.value_counts().index
country_names
```

```python
# Pie chart
plt.pie(country_values,labels=country_names)
```

```python
# Therefore, top 3 countries
plt.pie(country_values[:3],labels=country_names[:3], autopct='%1.1f%%')
```

##### Observation: 
Maximum transactions of zomato are from india. 
Then comes US and UK at 2nd and 3rd position respectively.

```python
# Categorize ratings

ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

### Observation
1. 4.5 to 4.9---> Excellent
2. 4.0 to 3.4--->very good
3. 3.5 to 3.9----> good
4. 3.0 to 3.4----> average
5. 2.5 to 2.9----> average
6. 2.0 to 2.4----> Poor

```python
ratings.head()
```

```python
import matplotlib

# makes diagram wider for clear text representations
# applies to each diagram in the code
matplotlib.rcParams['figure.figsize'] = (12,6) 

sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings)
```

```python
# map colors with name
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings, palette=['white','red','orange','yellow','green','green'])
```

```python
sns.countplot(x='Rating color',data=ratings, palette=['blue','red','orange','yellow','green','green'])
```
