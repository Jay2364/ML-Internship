```python
# numpy, pandas, seaborn, matplotlib

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

### Gaussian / Normal

```python
mu=0
sigma=0.1

s = np.random.normal(mu, sigma, size=10000)
```

```python
s
```

```python
plt.hist(s, 30)
plt.show()
```

```python
sns.distplot(s)
```

### Uniform distribution (discrete, continuous)

```python
# all have equal probability to occur in a range
u = np.random.uniform(0,1,10000)
```

```python
plt.hist(u)
plt.show()
```

```python
sns.distplot(u)
```

### Log normal

```python
l = np.random.lognormal(mean=0, sigma=0.6, size=1000)
l = l[np.where(abs(l) < 5)]
plt.hist(l,bins=100)
plt.show()
```

```python
# since lognormal is right skewed, outliers are present
sns.boxplot(l)
```

```python
sns.distplot(l)
```

```python
# lognormal -->> normal distribution
sns.distplot(np.log(l))
```

```python
sns.distplot(np.sqrt(l))
```

### Chi-square distribution

```python
c = np.random.chisquare(df = 30, size = 10000)
# c = c[np.where(abs(c) < 70)]
plt.hist(c, 100)
plt.show()
```

### Multi-chi-square

```python
d = np.random.chisquare(df=10, size=10000)
e = np.random.chisquare(df=3, size=10000)
c = c[np.where(abs(c) < 50)]
d = d[np.where(abs(d) < 50)]
e = e[np.where(abs(e) < 50)]
plt.hist(c, 100)
plt.hist(d, 100)
plt.hist(e, 100)
plt.show()
```

```python
f = np.random.chisquare(df=1, size=10000)
f = f[np.where(abs(f) < 50)]
plt.hist(c, 100)
plt.hist(d, 100)
plt.hist(e, 100)
plt.hist(f, 100)
plt.show()
```

### T-dist

```python
t = np.random.standard_t(df=3, size = 10000)
t = t[np.where(t < abs(10))]
plt.hist(t, 1000)
plt.show()
```

```python
u = np.random.standard_t(df=10, size = 10000)
v = np.random.standard_t(df=20, size = 10000)
w = np.random.standard_t(df=30, size = 10000)

u = u[np.where(u < abs(10))]
v = v[np.where(v < abs(10))]
w = w[np.where(w < abs(10))]

plt.hist(t, 100)
plt.hist(u, 100)
plt.hist(v, 100)
plt.hist(w, 100)
plt.show()
```

```python
plt.hist(w, 100)
plt.show()
```

### F-dist

```python
f = np.random.f(dfnum = 100, dfden = 100, size = 10000)
f = f[np.where(abs(f) < 10)]
plt.hist(f, 1000)
plt.show()
```

```python
g = np.random.f(dfnum = 10, dfden = 10, size = 10000)
h = np.random.f(dfnum = 4, dfden = 4, size = 10000)
i = np.random.f(dfnum = 1, dfden = 1, size = 10000)

g = g[np.where(abs(g) < 10)]
h = h[np.where(abs(h) < 10)]
i = i[np.where(abs(i) < 10)]

plt.hist(g, 1000)
plt.hist(h, 1000)
plt.hist(i, 1000)
plt.show()
```

## Discrete Dist

### Bernoulli

```python
# It is binomial with one trial, Therefore: 
b = np.random.binomial(n = 1, p = 0.7, size=1000)
plt.hist(b)
plt.show()
```

```python
bi = np.random.binomial(n = 30, 0 p = 0.72, size=1000)
plt.hist(bi)
plt.show()
```

### Poisson dist

```python
p = np.random.poisson(lam = 5, size = 1000)
plt.hist(p, 20)
plt.show()
```
